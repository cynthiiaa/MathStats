```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def prior(r, s, theta):
  return (1/(math.gamma(r)*math.gamma(s))) * math.gamma(r+s) * np.power(theta, r-1) * np.power(1-theta, s-1)

def posterior(r, s, theta, k, n):
  return (1/(math.gamma(r+k)*math.gamma(s+n-k))) * math.gamma(n+r+s) * np.power(theta, r+k-1) * np.power(1-theta, n+s-k-1)

theta = np.linspace(0,1,100)
```

Plot prior and posterior distributions for all 4 cases defined by the following values for parameters: (r,s) = (1,1),(4,4) & (n,k) = (4,3),(20,11)

```python
r = s = 1
n = 4
k = 3

prior_dist1 = prior(r,s,theta)
posterior_dist1 = posterior(r,s,theta,k,n)

r = s = 4
n = 4
k = 3

prior_dist2 = prior(r,s,theta)
posterior_dist2 = posterior(r,s,theta,k,n)

# graph stuff
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))
fig.suptitle('Prior and Posterior Distributions for n=4 and k=3')
ax1.set_title("(r,s)=(1,1)")
ax1.plot(theta, prior_dist1, label="prior")
ax1.plot(theta, posterior_dist1, label="posterior")
ax2.set_title("(r,s)=(4,4)")
ax2.plot(theta, prior_dist2, label="prior")
ax2.plot(theta, posterior_dist2, label="posterior", alpha=0.8)
ax1.set_xlabel('theta')
ax2.set_xlabel('theta')
ax2.set_ylabel('distribution')
ax1.set_ylabel('distribution')
ax1.legend()
ax2.legend()
plt.show()
```

For each of the 4 cases, determine the probability that the coin is biased towards heads i.e. $$ P(Î¸>0.5) $$

```python
from scipy.stats import beta # using beta distribution to find the probabilities

# Case 1
r = s = 1
n = 4
k = 3
a = r + k
b = n + s - k
probability = 1 - beta.cdf(0.5, a, b)
print("Case 1: ", probability)

# Case 2
r = s = 1
n = 20
k = 11
a = r + k
b = n + s - k
probability = 1 - beta.cdf(0.5, a, b)
print("Case 2: ", probability)

# Case 3
r = s = 4
n = 4
k = 3
a = r + k
b = n + s - k
probability = 1 - beta.cdf(0.5, a, b)
print("Case 3: ", probability)

# Case 4
r = s = 4
n = 20
k = 11
a = r + k
b = n + s - k
probability = 1 - beta.cdf(0.5, a, b)
print("Case 4: ", probability)
```
